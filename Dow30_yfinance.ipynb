```python

# Define list of ticker symbols for which you want to pull data
tickers = ["MMM",
"AMZN",
"AXP",
"AMGN",
"AAPL",
"CAT",
"CVX",
"CSCO",
"DOW",
"GSBD",
"HON",
"INTC",
"IBM",
"JPM",
"JNJ",
"MCD",
"MRK",
"MSFT",
"NKE",
"BA",
"KO",
"HD",
"PG",
"TRV",
"DIS",
"UNH",
"VZ",
"V",
"WMT",
"CRM"
]


# Create empty list to store the data
data = []

# Fetch data for each ticker symbol
for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)
        info = stock.info
        pe_ratio = info.get('trailingPE', None)   # 'None' is a Python keyword used as a default return value to indicate the absence of a value or when the value is not available
        data.append({
            'Symbol': ticker,
            'Company (long) name': info.get('longName', None),            
            'Sector': info.get('sector', None),
            'Industry': info.get('industry', None),
            'Current price': info.get('currentPrice', None),
            'Debt-to-equity': info.get('debtToEquity',None),
            'Trailing P/E': pe_ratio,
            'Forward P/E': info.get('forwardPE', None),
            '52-week low': info.get('fiftyTwoWeekLow', None),
            '52-week high': info.get('fiftyTwoWeekHigh', None),
            '52-week change': info.get('52WeekChange', None),
            'Shares outstanding': info.get('sharesOutstanding', None),
            'PEG ratio': info.get('pegRatio', None),
            'Price-to-book': info.get('priceToBook', None),
            'Return on equity': info.get('returnOnEquity', None),
            'Trailing Price-to-sales': info.get('priceToSalesTrailing12Months', None),
            'Average volume': info.get('averageVolume', None),
            'Market cap': info.get('marketCap', None),
            'Full-time Employees': info.get('fullTimeEmployees', None)
        })
    except Exception as e:
        print(f"Failed to retrieve data for {ticker}. Error: {e}")

# Create pandas dataframe from the list of dictionaries
df = pd.DataFrame(data)
df
```

```python
df.to_csv("Dow30_yfinance_Q2.csv", header=True, index=False)
```
